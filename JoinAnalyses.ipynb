```python
import pandas as pd
```

```python
homepage_names = ["homepage_data_analyzed_multiple_2.json", "homepage_data_analyzed_multiple_3.json", "homepage_data_analyzed_multiple_4.json", "homepage_data_analyzed_multiple_5.json"]
number_counter = 0
joined_df = pd.DataFrame()

model_name = "gpt-4o"
new_dfs = []

for homepage_index, homepage_name in enumerate(homepage_names):
    homepage_data = pd.read_json(homepage_name)
    new_df = pd.DataFrame()
    for column in homepage_data.columns:
        new_name = column + "_" + str(homepage_index)
        new_df[new_name] = homepage_data[column]
    new_df["name"] = homepage_data["name"]
    new_dfs.append(new_df)

joined_df = new_dfs[0]
for df in new_dfs[1:]:
    joined_df = joined_df.merge(df, on="name", how="outer")
joined_df.head()
```

```python
general_analysis_columns = [col for col in joined_df.columns if 'general_analysis' in col]
num_general_analysis_columns = len(general_analysis_columns)
print("Number of columns with 'general_analysis' in their name:", num_general_analysis_columns)
```

```python
joined_df.to_json(f"homepage_data_analyzed_multiple_joined_{num_general_analysis_columns}.json")
print("Total number of columns:", len(joined_df.columns))
```
